#### 1) LIST

```python
mahasiswa = ["Iin", "Darma", "Aca", "Lutfi"]
mahasiswa.append("Elsa")
mahasiswa.insert(3, "Ami")
mahasiswa.pop()
mahasiswa.sort(reverse=True)
mahasiswa.remove("Aca")

i = 1

for nama in mahasiswa:
    print("Nama Mahasiswa ke -", i ,": ", nama)
    i += 1

print(mahasiswa)
```

##### Analisis

Kode ini menunjukkan bagaimana cara mengubah isi list di Python. Program menambah nama “Elsa”, lalu menyisipkan “Ami” di posisi ke-3. Setelah itu, elemen terakhir dihapus menggunakan pop(). Data kemudian diurutkan dari huruf terbesar ke terkecil dengan sort(reverse=True), dan nama “Aca” dihapus memakai remove(). Bagian perulangan digunakan untuk menampilkan seluruh nama mahasiswa satu per satu dengan penomoran otomatis. Pada akhirnya, list yang telah diperbarui ditampilkan sebagai hasil akhir.

```python
mahasiswa = ["Iin", "Aca", "Ami", "Lutfi", "Adit"]
print(mahasiswa)
```

##### Analisis

Kode ini menyimpan beberapa nama mahasiswa ke dalam list. Saat program dijalankan, semua nama di dalam list ditampilkan dalam satu baris sesuai urutan yang sudah ditentukan.

#### 2) WHILE LOOP

```python
i = 1
while i <= 13:
    if i == 7:
        print("WOW")
    else:
        print(i)
    i += 2
```

##### Analisis

Kode ini memakai perulangan while untuk menampilkan bilangan ganjil dari 1 sampai 13. Setiap kali loop dijalankan, nilai i naik dua. Ketika mencapai angka 7, program menampilkan kata “WOW” sebagai pengganti angka tersebut. Setelah itu, loop berlanjut hingga batas akhir terpenuhi.

```python
i = 0
while i <= 5:
    print(i)
    i += 1
```

##### Analisis

Kode ini menampilkan angka dari 0 sampai 5 dengan bantuan perulangan while. Selama nilai i belum lebih besar dari 5, program akan mencetak nilainya lalu menambahkannya satu per satu. Setelah i mencapai 6, kondisi while menjadi salah sehingga perulangan berhenti.

#### 3) SET

```python
mahasiswa = {"Aca", "Lutfi", "Ami", "Iin", "Iin", "Elsa"}
mahasiswa [0] = "Adit"
print(mahasiswa)
```

##### Analisis

Kode program tersebut menampilkan pesan error karena Python mengenali data mahasiswa sebagai set, bukan daftar (list). Berbeda dengan list, set tidak memiliki urutan atau indeks, sehingga elemennya tidak bisa diakses satu per satu menggunakan angka indeks. Ketika perintah mahasiswa[0] = "Adit" dijalankan, Python menolak karena operasi semacam itu tidak berlaku pada set.

#### 4) DICTIONARY

```python
person = {"name": "Nur Insyirah AR", "Hp": "9418", "Kelas": "TEKOM F"}
print(person)
```

##### Analisis

Kode ini membuat data seseorang dalam bentuk dictionary. Di dalamnya terdapat tiga data, yaitu nama, nomor HP, dan kelas. Ketika print(person) dijalankan, Python akan menampilkan seluruh isi dictionary dalam satu baris.

```python
person = {"name": "Nur Insyirah AR", "Hp": "9418", "Kelas": "TEKOM F"}
for key, value in person.items():
    print(key, ":", value)
```

##### Analisis

Kode ini menyimpan data seseorang dalam dictionary person, lalu menampilkan isinya satu per satu menggunakan perulangan for. Fungsi items() digunakan agar program bisa mengambil setiap pasangan key dan value dari dictionary. Saat dijalankan, program akan mencetak nama atribut (key) dan isinya (value) dalam baris terpisah.

#### 5) FOR LOOP

```python
for i in range(10):
  print(i)
```

##### Analisis

Kode ini menampilkan urutan angka dari 0 sampai 9. Fungsi range(10) menentukan bahwa perulangan berhenti sebelum mencapai 10, dan perulangan for akan mencetak satu angka setiap kali loop berjalan hingga selesai.

```python
for i in range(1,6):
  print(i)
```

##### Analisis

Kode ini menampilkan angka 1 sampai 5 secara berurutan. Fungsi range(1,6) menentukan bahwa perulangan dimulai dari angka 1 dan berhenti sebelum mencapai 6. Setiap angka dicetak satu per satu menggunakan print().

```python
for i in range(5,10):
  print(i)
```

##### Analisis

Kode ini menampilkan angka 5 sampai 9 secara berurutan. Fungsi range(5,10) menentukan bahwa perulangan dimulai dari angka 5 dan berhenti sebelum mencapai 10. Setiap angka dicetak satu per satu menggunakan print().

#### LANJUTAN LIST

```python
thislist = ["Apple", "Banana", "Cherry", "Orange", "Kiwi", "Melon", "Mango"]
print(thislist[2:3])
```

##### Analisis

Kode ini membuat list berisi beberapa nama buah, lalu menampilkan sebagian datanya menggunakan teknik slicing. Potongan [2:3] berarti mengambil elemen dari indeks 2 sampai sebelum 3, sehingga hasil yang ditampilkan adalah ["Cherry"].

```python
thislist = ["Apple", "Banana", "Cherry", "Orange", "Kiwi", "Melon", "Mango"]
print(thislist[3:])
```

##### Analisis

Kode ini menyimpan beberapa nama buah di dalam list. Perintah [3:] berarti mengambil semua elemen mulai dari indeks ke-3 sampai elemen terakhir. Dengan begitu, hasil yang muncul di layar adalah ["Orange", "Kiwi", "Melon", "Mango"].

```python
thislist = ["Apple", "Banana", "Cherry", "Orange", "Kiwi", "Melon", "Mango"]
print(thislist[:])
```

##### Analisis

Kode ini menyimpan beberapa nama buah di dalam list. Ketika program menuliskan [:], artinya Python menampilkan semua isi list dari awal sampai akhir tanpa menghilangkan satu pun elemen. Jadi hasil akhirnya sama seperti isi list aslinya.

```python
thislist = ["Apple", "Banana", "Cherry", "Orange", "Kiwi", "Melon", "Mango"]
print(thislist[1:3])
```

##### Analasis

Kode di atas menyimpan beberapa nama buah dalam list. Saat Python membaca thislist[1:3], artinya hanya elemen mulai dari posisi kedua sampai sebelum posisi keempat yang akan ditampilkan. Karena indeks dihitung dari nol, maka hasilnya adalah "Banana" dan "Cherry".

#### LANJUTAN WHILE LOOP

```python
i = 1
while i <= 9:
    print(i)
    i += 2
```

##### Analisis

Kode ini mencetak bilangan ganjil dari 1 sampai 9. Nilai i dimulai dari 1 dan setiap kali perulangan berjalan, nilainya bertambah 2. Perulangan berhenti ketika i lebih dari 9. Hasil akhirnya menampilkan angka 1, 3, 5, 7, dan 9 di layar.

```python
i = 1
while i <= 13:
    if i == 7:
        print("WOW")
    else:
        print(i)
    i +=2
```

##### Analisis

Kode ini menampilkan deretan angka ganjil dari 1 sampai 13. Saat angka 7 muncul, program menggantinya dengan tulisan “WOW”. Setelah itu, perulangan tetap berlanjut hingga angka 13.

```python
i = 1
while i <= 8:
    if i % 2 == 1:
        print("hidup")
    else:
        print("mahasiswa")
    i += 1
```

##### Anaisis

Kode ini membuat tulisan “hidup” dan “mahasiswa” muncul secara selang-seling. Nilai i dimulai dari 1 dan naik terus hingga 8. Jika i ganjil, Python menulis “hidup”; kalau genap, Python menulis “mahasiswa”. Hasil akhirnya menampilkan pola hidup–mahasiswa secara bergantian.

#### LANJUTAN FOR LOOP

```python
for i in range(1, 14):
    print(i)
    if i % 3 == 2:  
        print("Ha")
```

##### Analisis

Kode ini menampilkan angka dari 1 sampai 13 secara berurutan. Setelah mencetak setiap angka, program memeriksa apakah angka tersebut memiliki sisa 2 jika dibagi 3. Kalau iya, maka tulisan “Ha” ditampilkan tepat setelahnya. Akibatnya, kata “Ha” muncul di bawah angka 2, 5, 8, dan 11.

```python
for i in range(1, 14):
    if i % 3 == 0:
        continue
    print(i)
    if i % 3 == 2:
        print("Ha")
    if i % 6 == 4:
        print("Hi")
```

##### Analisis

Kode ini menampilkan angka dari 1 sampai 13, tapi melewatkan angka yang bisa dibagi 3. Setelah itu, setiap kali angka memiliki sisa 2 saat dibagi 3, program menulis “Ha”. Kalau angkanya punya sisa 4 saat dibagi 6, program menulis “Hi”. Jadi “Ha” muncul di 2, 5, 8, dan 11, sedangkan “Hi” di 4 dan 10.
