## 23. Python Arithmetic Operators

```python
x = 15
y = 4

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x ** y)
print(x // y)
```

### Analisis

Kode tersebut melakukan berbagai operasi aritmatika antara variabel x = 15 dan y = 4.
Berikut penjelasan setiap operasi dan hasilnya:
- x + y → penjumlahan → 15 + 4 = 19
- x - y → pengurangan → 15 - 4 = 11
- x * y → perkalian → 15 × 4 = 60
- x / y → pembagian → 15 ÷ 4 = 3.75
- x % y → sisa bagi (modulus) → sisa dari 15 ÷ 4 = 3
- x ** y → pemangkatan → 15⁴ = 50625
- x // y → pembagian bulat (floor division) → hasil bagi tanpa desimal = 3

## 24. Python Assignment Operators

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

### Analisis

Kode tersebut menggunakan dua cara untuk memeriksa jumlah elemen dalam list numbers = [1, 2, 3, 4, 5].
- Pada bagian pertama, fungsi len(numbers) menghitung jumlah elemen (5) dan menyimpannya ke variabel count. Karena count > 3, program menampilkan "List has 5 elements".
- Pada bagian kedua, digunakan walrus operator (:=) yang memungkinkan penugasan nilai di dalam ekspresi kondisi. Di sini, count langsung diisi dengan hasil len(numbers) di dalam pernyataan if, tanpa harus dideklarasikan terpisah.

## 25. Python Comparison Operators

```python
x = 5
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

### Analisis

Kode tersebut menggunakan operator perbandingan untuk membandingkan nilai antara variabel x = 5 dan y = 3. Setiap pernyataan menghasilkan nilai logika True atau False sesuai kondisi berikut:
- x == y → 5 sama dengan 3 → False
- x != y → 5 tidak sama dengan 3 → True
- x > y → 5 lebih besar dari 3 → True
- x < y → 5 lebih kecil dari 3 → False
- x >= y → 5 lebih besar atau sama dengan 3 → True
- x <= y → 5 lebih kecil atau sama dengan 3 → False

## 37. Python Nested If

```python
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

### Analisis

Kode tersebut menggunakan percabangan bersarang (nested if), yaitu pernyataan if di dalam if lain.
- Pertama, kondisi x > 10 bernilai True karena 41 > 10, maka program menampilkan "Above ten,".
- Di dalamnya ada if kedua yang memeriksa x > 20, dan hasilnya juga True karena 41 > 20, sehingga program menampilkan "and also above 20!".

## 38. Python Pass Statement

```python
a = 33
b = 200

if b > a:
  pass
```

### Analisis

## 40. Python Match

```python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
```

### Analisis

Kode tersebut menggunakan struktur kontrol match-case, fitur baru di Python 3.10 yang mirip dengan switch-case pada bahasa lain. Nilai variabel day = 4 akan dibandingkan dengan setiap case yang ada. karena nilai day sama dengan 4, maka output yang ditampilkan adalah "Thursday".
